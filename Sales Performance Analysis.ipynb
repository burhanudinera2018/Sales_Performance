```python
# Sekalian menghapus kolom 'Unnamed: 9' dan 'Unnamed: 10'
sales = pd.read_csv('/Users/mac/Study/Study_burhanudin/Common_Data/100_Sales.csv').drop(columns=['Unnamed: 9','Unnamed: 10'])
```

# Data Analysis

+ Tujuan dari analitik penjualan ini adalah untuk menyederhanakan informasi yang tersedia bagi perusahaan dan membantu mereka memahami dengan jelas kinerja penjualan, tren penjualan, dan peluang.

```python
# Create definisi variable waktu

days = sales['Ship_Date'].str.split('/').str.get(0).values
months = sales['Ship_Date'].str.split('/').str.get(1).values
years = sales['Ship_Date'].str.split('/').str.get(2).values
```

```python
# Definisi tahun

date_series = years +'/' + months + '/'+ days
```

```python
# merubah format waktu pada kolom 'ship_date' ke date_series

sales['Ship_Date'] = pd.to_datetime(date_series)
```

```python
#notice the memory usage before conveting the data types

sales.info()
```

```python
# Mari kita periksa variabel yang paling cocok untuk kategori dtype

sales.nunique()
```

```python
# Mendefinisikan data type ke dalam bentuk 'category'

sales['Region'] = sales['Region'].astype('category')
sales['Item_Type'] = sales['Item_Type'].astype('category')
sales['Sales_Channel'] = sales['Sales_Channel'].astype('category')
sales['Order_Priority'] = sales['Order_Priority'].astype('category')
```

```python
# Kembali di periskas data type

sales.info() 
```

```python
# Melihat kembali sekilas data

sales.head() #take a look at data frame structure
```

# Memahami variabel yang ada dan membuat variabel baru

**Revenue** adalah jumlah total pendapatan yang dihasilkan oleh penjualan barang atau jasa yang terkait dengan operasi utama perusahaan.

**Profit** biasanya disebut laba bersih atau garis bawah, adalah jumlah pendapatan yang tersisa setelah memperhitungkan semua biaya, utang, aliran pendapatan tambahan, dan biaya operasi.

Jadi jika kita ingin mendapatkan biayanya kita bisa mendapatkannya dengan cara mengurangkan Laba dari Pendapatan tersebut.

```python
# Menanmbahkan colum baru 'Sales[Expense]' sesuai rumus :

sales['Expense'] = sales['Total_Revenue'] -sales['Total_Profit']
```

**Profit Margin** adalah rasio keuntungan perusahaan (penjualan dikurangi semua biaya) dibagi dengan pendapatannya. 

+ Rasio margin laba membandingkan laba dengan penjualan dan memberi tahu Anda seberapa baik perusahaan menangani keuangannya secara keseluruhan. Itu selalu dinyatakan sebagai persentase.

```python
# Profit Margin

sales['Profit_Margin'] = (sales['Total_Profit'] / sales['Total_Revenue'] ) * 100
```

```python
# Lihat sekilas dataset

sales.head()
```

+ Seperti yang kita ketahui sekarang bahwa Profit Margin banyak bercerita tentang kinerja perusahaan, mari kita rangking berdasarkan Profit Margin mereka.

```python
# Wilayah mana yang menghasilkan Margin Keuntungan tertinggi? * dan itu negara terkait?

sales.sort_values(by='Profit_Margin', ascending=False).head(1)
```

```python
# Wilayah mana yang menghasilkan Laba tertinggi? * dan itu negara terkait?

regions = sales.groupby(by='Region')
```

```python
df_highest_profit_regions = pd.DataFrame(columns=sales.columns)
```

```python
# Definisi Fungsi

for reg, data in regions:
    highest_profit = data.nlargest(1, 'Total_Profit')
    df_highest_profit_regions = pd.concat([df_highest_profit_regions, highest_profit])
```

```python
df_highest_profit_regions.sort_values(by='Total_Profit', ascending=False)
```

```python
# Visualized Profit ke 7 Region dengan memasukan nilainya berdasarkan runing scrip python di atas

fig, ax = plt.subplots(figsize=(10,7))

tmp_df = df_highest_profit_regions.copy()

tmp_df = tmp_df.sort_values(by='Total_Profit', ascending=False)

sns.barplot(x = tmp_df['Total_Profit'], y=tmp_df['Region'], orient='h', data=tmp_df,\
           order=['Middle East and North Africa', 'Australia and Oceania', 'Europe', 'Central America and the Caribbean', 'Asia', 'Sub_Saharan Africa', 'North America'],
            capsize=0.2, ax=ax)

lbs = ['Middle East and North Africa: 1719922.04 \n Country: Pakistan', 
      'Australia and Oceania: 1678540.98 \n Country: Samoa',
      'Europe: 1541705.29 \n Country: Iceland',
      'Central America and the Caribbean: 1487261.02 \n Country: Honduras',
      'Asia: 1367272.50 \n Country: Myanmar',
      'Sub_Saharan Africa: 1254472.05 \n Country: Djibouti',
      'North America: 1152486.42 \n Country: Mexico']

ax.bar_label(ax.containers[-1], labels=lbs,label_type='center')

plt.title('Region with their Profits')
plt.xlabel('Regions')
plt.ylabel('Total Profit')
plt.show()
```

# Jenis Item mana yang memiliki Total Pendapatan tertinggi?

```python
# Create function

def highlight_cols(x): 
    df = x.copy()
    df.loc[:, 'Total_Revenue'] = 'background-color: green'
    df[['Country', 'Unit_Cost', 'Total_Profit','Expense']] = 'background-color: grey'
    return df 
```

```python
display(tmp_df.style.apply(highlight_cols, axis = None))
```

```python
# Reorder the dataframe
df = tmp_df['Total_Revenue'].to_frame()
df = df.sort_values(by=['Total_Revenue'])

# initialize the figure
plt.figure(figsize=(30,15))
ax = plt.subplot(111, polar=True)
plt.axis('off')

# Constants = parameters controling the plot layout:
upperLimit = 100
lowerLimit = 30
labelPadding = 4

# Compute max and min in the dataset
max = df['Total_Revenue'].max()

# Let's compute heights: they are a conversion of each item value in those new coordinates
# In our example, 0 in the dataset will be converted to the lowerLimit (10)
# The maximum will be converted to the upperLimit (100)
slope = (max - lowerLimit) / max
heights = slope * df['Total_Revenue'] + lowerLimit

# Compute the width of each bar. In total we have 2*Pi = 360°
width = 2*np.pi / len(df.index)

# Compute the angle each bar is centered on:
indexes = list(range(1, len(df.index)+1))
angles = [element * width for element in indexes]
angles

# Draw bars
bars = ax.bar(
    x=angles, 
    height=heights, 
    width=width, 
    bottom=lowerLimit,
    linewidth=2, 
    edgecolor="white",
    color="#61a4b2",
)

# Add labels
for bar, angle, height, label in zip(bars,angles, heights, df.index):

    # Labels are rotated. Rotation must be specified in degrees :(
    rotation = np.rad2deg(angle)

    # Flip some labels upside down
    alignment = ""
    if angle >= np.pi/2 and angle < 3*np.pi/2:
        alignment = "right"
        rotation = rotation + 180
    else: 
        alignment = "left"

    # Finally add the labels
    ax.text(
        x=angle, 
        y=lowerLimit + bar.get_height() + labelPadding, 
        s=label, 
        ha=alignment, 
        va='center', 
        rotation=rotation, 
        rotation_mode="anchor") 
```
